```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv/tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
# 20 columns in movies dataset and 4 columns in credits dataset
print(movies.shape)
print(credits.shape)
```

```python
# merging movies and credits dataset on title column
movies = movies.merge(credits, on = 'title')
print(movies.shape)
```

```python
movies.head(1)
```

```python
# dropping unwanted columns
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_companies
# production_countries
# release_date
# runtime
```

```python
# checking no of languages major languages are en so we can drop this column
print(movies['original_language'].value_counts())
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(1)
```

```python
# checking for null values
movies.isnull().sum()
```

```python
# dropping null values
movies.dropna(inplace = True)
movies.isnull().sum()
```

```python
import ast
```

```python
movies.head(1)
```

```python
import ast
```

```python
# creating a function to convert stringified list of dictionaries to list of names

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# fixing genres column
movies['genres'] = movies['genres'].apply(convert)
movies.head(1)
```

```python
movies['keywords'][0]
```

```python
# fixing keywords column
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

```python
movies['cast'][0]
```

```python
# fixing cast column need top 3 cast members
# step 1 fixing cast
movies['cast'] = movies['cast'].apply(convert)
```

```python
# selecting top 3 cast members
movies['cast'] = movies['cast'].apply(lambda x:x[:3])
```

```python
movies.head(1)
```

```python
# creating a fucntion to fetch director name from crew column
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
# check data types of columns
movies.dtypes
```

```python
# converting overview to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

```python
# convert names like "saqib ahmed" → "saqibahmed" inside a list
def remove_spaces(lst):
    return [i.replace(" ", "") for i in lst]
```

```python
movies['genres'] = movies['genres'].apply(remove_spaces)

movies['keywords'] = movies['keywords'].apply(remove_spaces)

movies['cast'] = movies['cast'].apply(remove_spaces)

movies['crew'] = movies['crew'].apply(remove_spaces)
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

movies.head(1)
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)).head()
```

```python
new_df['tags'][0]
```

```python
# converting to lower case
new_df['tags'] = new_df['tags'].fillna('').str.lower()

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

euclidean distance is not used in higher dimmensions because of curse of dimmensionality. In higher dimmensions, the distance between points becomes less meaningful and can lead to inaccurate results. Instead, other distance metrics such as cosine similarity or Manhattan distance are often used in higher dimmensions for tasks like recommendation systems.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
